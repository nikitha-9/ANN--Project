````python
import tensorflow as tf
import torch
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
(x_train,y_train),(x_test,y_test)=data.load_data()
````

````python
# flattening x_train and x_test
flat_x_train=x_train.reshape(len(x_train),28*28)
flat_x_test=x_test.reshape(len(x_test),28*28)
````

````python
flat_x_train.shape
````

#keras sequential units(neuron)

````python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid'),
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
````

````python
model.fit(flat_x_train,y_train,epochs=5)
````

````python
tf.config.run_functions_eagerly(True)
````

````python
new_x_train=x_train/255
new_x_test=x_test/255
````

````python
flat_x_train=new_x_train.reshape(len(new_x_train),28*28)
flat_x_test=new_x_test.reshape(len(new_x_test),28*28)
````

````python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid'),
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
````

````python
model.fit(flat_x_train,y_train,epochs=5)
````

````python
model.evaluate(flat_x_test,y_test)
````

````python
y_pred=model.predict(flat_x_test)
````

````python
y_pred[0]
````

````python
np.argmax(y_pred[0])
````

````python
y_pred_label=[np.argmax(i) for i in y_pred]
````

````python
y_pred_label
````

````python
# confusion matrix
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred_label)
````

````python
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('actual')
plt.ylabel('predicted')
````

````python
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid'),
    keras.layers.Dense(100,activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
````

````python
model.fit(flat_x_train,y_train,epochs=5)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Choose an index from the test set
index = 0

# Get the image and its true label
image = x_test[index]
true_label = y_test[index]

# Reshape the image for prediction
image_reshaped = image.reshape(1, 28 * 28) / 255.0

# Make prediction using the trained model
prediction = model.predict(image_reshaped)
predicted_label = np.argmax(prediction)

# Display the image and prediction
plt.imshow(image, cmap='gray')
plt.title(f"True Label: {true_label}, Predicted Label: {predicted_label}")
plt.axis('off')
plt.show()
````
